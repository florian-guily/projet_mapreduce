```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
measures = pd.read_csv("measures.csv", index_col='method')
measures = measures.T
seq = measures['Sequentiel']
measures.drop('Sequentiel', axis=1, inplace=True)
x = [1, 2, 3, 4, 5, 8, 10]
full_measures = measures[measures.columns[::2]]
AS_measures = measures[measures.columns[1::2]]

fm_mean = full_measures.mean()
am_mean = AS_measures.mean()
fm_mean.index = x
am_mean.index = x

plt.figure(figsize=(15,8))
fm_mean.plot(label="Distribué avant envoie")
am_mean.plot(label="Distribué après envoie")
plt.plot(x, [seq.mean()for i in x], label="Séquentiel")
plt.xlabel("Nombre de noeud")
plt.ylabel("Temps d'exécution (s)")
plt.title("Temps d'exécution du wordcount sur un fichier de 300Mo en fonction du nombre de noeud")
plt.legend()
plt.show()
#seq.mean().plot()
```
